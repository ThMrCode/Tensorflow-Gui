```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

**L_total = L_original + λ * Regularización**

`L_total` es la pérdida total que el optimizador tratará de minimizar `L_original` es la pérdida original (como la entropía cruzada, MSE, etc.). `λ` es el parámetro de regularización (también conocido como coeficiente de penalización). `Regularización` puede ser una penalización L1 (suma de los valores absolutos de los pesos) o L2 (suma de los cuadrados de los pesos).

**W = W - α * (G + λ * R)**

`W` son los pesos del modelo. `α` es la tasa de aprendizaje (learning rate). `G = ∂L_original / ∂W` es el gradiente de la pérdida original con respecto a los pesos. `λ` es el coeficiente de regularización. `R = ∂Regularización / ∂W` es el gradiente del término de regularización.

**Regularizacion (L1 (Lasso), L2 (Ridge))**

  **A los pesos L1 = λ1∑∣w∣, L2 = λ2∑w2**

  **A los bias  L1 = λ1∑∣b∣, L2 = λ2∑b2**

  **A la activacion  L1 = λ1∑∣activation∣, L2 = λ2∑activation2** Esto afecta a capas anteriores

```python
l1 = tf.keras.regularizers.L1(0.01)
l2 = tf.keras.regularizers.L2(0.01)
l1l2 = tf.keras.regularizers.L1L2(0.01,0.01)
ortogonal = tf.keras.regularizers.OrthogonalRegularizer(0.01,'rows')
print(l1, "\n", l2, "\n", l1l2, "\n", ortogonal)
```
