```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

**Tensorflow usa Autodiferenciaci√≥n**

**Rastrea todas las operaciones hechas sobre los tensores, en un grafo computacional, y asi obtiene los gradientes correspondientes**

```python
conv2d = tf.keras.layers.Conv2D(
    filters=1,
    kernel_size=(3,3),
    strides=(1,1),
    padding='same',
    data_format='channels_last',
    dilation_rate=(1,1),
    groups=1,
    activation=None,
    use_bias=True,
    kernel_initializer= tf.keras.initializers.RandomNormal(),
    bias_initializer= tf.keras.initializers.Zeros(),
    kernel_regularizer=None,
    bias_regularizer=None,
    activity_regularizer=None,
    kernel_constraint=None,
    bias_constraint=None,
)
```

```python
# HYPERPARAMETERS
inputShape = tf.TensorShape([1,3,3,1])
outputShape = conv2d.compute_output_shape(inputShape)
conv2d.build(input_shape=inputShape)
```

```python
inputTensor = tf.Variable(tf.random.normal(inputShape,mean=0.0,stddev=1.0))
outputTensor = conv2d.call(inputTensor)
```

```python
initKernel = conv2d.get_weights()[0]
initBias = conv2d.get_weights()[1]
print("INPUT \n", inputTensor[0,:,:,0].numpy())
print("OUTPUT \n", outputTensor[0,:,:,0].numpy())
print("KERNEL \n", conv2d.get_weights()[0][:,:,0,0])
print("BIAS", conv2d.get_weights()[1])
```

```python
conv2d.set_weights([tf.Variable(tf.ones_like(initKernel)),initBias])
outputTensor = conv2d.call(inputTensor)
```
